# sampling from u(0,1), numbers

```python
sample_size = 500
rn1 = npr.rand(sample_size, 3)     # 生成 500 x3 的 uniform(0,1)亂數 
rn2 = npr.randint(0, 10, sample_size)   # 從 [0,10) 中生成 500 個整數亂數 
rn3 = npr.sample(size = sample_size)    # npr.sample 似 npr.rand 但只能一維, 恐龍級寫法  
a = [0, 25, 50, 75, 100]
rn4 = npr.choice(a, size=sample_size)  # 在 1-D list a 中抽出放回 sample_size 次 
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 8)) 
# plt.subplots 告訴 Python 擺成幾列幾行，2列2行，整體為 10x8英吋大小 
# ((ax1, ax2), (ax3, ax4)) 開 2x2 的陣列, 個別陣列 ax1 至 ax4 內容由下給出
ax1.hist(rn1, bins=25, stacked=True)  # rn1 是 500x3, 畫直方圖 25 bars , stacked=True 表堆疊 3 層 
ax1.set_title('rand')
ax1.set_ylabel('frequency')
ax1.grid(True)       # 格線 
ax2.hist(rn2, bins=25)   # rn2 是從 [0,10) 中生成 500 個整數亂數, 畫直方圖 25 bars 
ax2.set_title('randint')
ax2.grid(True)
ax3.hist(rn3, bins=25)  # rn3 是 npr.sample 似 npr.rand 但只能一維, 畫直方圖 25 bars
ax3.set_title('sample')
ax3.set_ylabel('frequency')
ax3.grid(True)
ax4.hist(rn4, bins=25)   # rn4 是在 1-D list a 中抽出放回 sample_size 次
ax4.set_title('choice')
ax4.grid(True)
```

# sampling from distributions

```python
sample_size = 100
rn1 = npr.standard_normal(sample_size)
rn2 = npr.normal(100, 20, sample_size)
rn3 = npr.chisquare(df=0.5, size=sample_size)
rn4 = npr.poisson(lam=1.0, size=sample_size)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
ax1.hist(rn1, bins=25)
ax1.set_title('standard normal')
ax1.set_ylabel('frequency')
ax1.grid(True)
ax2.hist(rn2, bins=25)
ax2.set_title('normal(100, 20)')
ax2.grid(True)
ax3.hist(rn3, bins=25)
ax3.set_title('chi square, df=0.5')
ax3.set_ylabel('frequency')
ax3.grid(True)
ax4.hist(rn4, bins=25)
ax4.set_title('Poisson, lam=1')
ax4.grid(True)
```

# generate normal random numbers

```python
import statistics
from scipy.stats import norm
x_axis = np.arange(-4,4,0.01)
mean=statistics.mean(x_axis)
sd=statistics.stdev(x_axis)
plt.plot(x_axis, norm.pdf(x_axis, mean, sd))
plt.show()
plt.plot(x_axis, norm.pdf(x_axis, 0, 1))
plt.show()
```

```python
import statistics
from scipy.stats import norm
x_axis=np.arange(-4,4,0.01)


sample_size = 250000
rn1 = npr.standard_normal(sample_size)

fig, ((ax1, ax2)) = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
ax1.plot(x_axis, norm.pdf(x_axis, 0, 1))
ax1.set_title('standard normal')
ax2.hist(rn1, bins=50)
ax2.set_title('n=250,000')
ax2.set_ylabel('frequency')
```
